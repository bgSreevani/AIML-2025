```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/MilkQuality.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Grade'] = le.fit_transform(df['Grade'])
df.head()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print("\nDataset Information:")
print(df.info())
print("\nFirst few rows of the dataset:")
print(df.head())

```

```python
grade_mapping = {'low': 0, 'medium': 1, 'high': 2}
df['Grade_numeric'] = df['Grade'].map(grade_mapping)

```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Grade', data=df, palette="viridis")
plt.title('Distribution of Milk Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
correlations_with_target = df.corr(numeric_only=True)
correlations_with_target = df.corr(numeric_only=True)['Grade_numeric'].sort_values(ascending=False)
print("\nCorrelation of Features with Grade (numeric):")
print(correlations_with_target)
```
